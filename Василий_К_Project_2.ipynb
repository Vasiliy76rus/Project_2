```python
# текст запроса
import psycopg2

# Подключение к базе данных
connection = psycopg2.connect(
    dbname='project_sql',
    user='skillfactory',
    password='cCkxxLVrDE8EbvjueeMedPKt',
    host='84.201.134.129',
    port=5432
)

# Создание курсора
cursor = connection.cursor()

# Выполнение запроса
query = '''
SELECT areas.name, COUNT(vacancies.id) AS cnt
FROM vacancies
JOIN areas ON vacancies.area_id = areas.id
GROUP BY areas.name
ORDER BY cnt DESC;
'''

cursor.execute(query)

# Получение результатов
results = cursor.fetchall()

# Закрытие соединения
cursor.close()
connection.close()

# Вывод результатов
for row in results:
    print(f"Регион: {row[0]}, Количество вакансий: {row[1]}")
```

2. Напишите запрос, который для каждого региона выведет количество работодателей и вакансий в нём.
Среди регионов, в которых нет вакансий, найдите тот, в котором наибольшее количество работодателей.


```python
# текст запроса
import psycopg2

# Подключение к базе данных (убрали повторное подключение)
conn = psycopg2.connect(
    dbname='project_sql',
    user='skillfactory',
    password='cCkxxLVrDE8EbvjueeMedPKt',
    host='84.201.134.129',
    port=5432
)

cur = conn.cursor()

# Первый запрос: количество работодателей и вакансий по регионам
sql_query_1 = '''
SELECT 
    a.name AS region_name,
    COUNT(DISTINCT e.id) AS employer_count,
    COUNT(v.id) AS vacancy_count
FROM 
    areas a
LEFT JOIN 
    employers e ON a.id = e.area
LEFT JOIN 
    vacancies v ON e.id = v.employer_id
GROUP BY 
    a.name
ORDER BY 
    region_name;
'''

cur.execute(sql_query_1)
results_1 = cur.fetchall()

# Для наглядности можно вывести результаты первого запроса
print("Количество работодателей и вакансий по регионам:")
for row in results_1:
    print(f"Регион: {row[0]}, Работодатели: {row[1]}, Вакансии: {row[2]}")

# Второй запрос: регион с наибольшим количеством работодателей без вакансий
sql_query_2 = '''
WITH region_employer_counts AS (
    SELECT 
        a.name AS region_name,
        COUNT(DISTINCT e.id) AS employer_count
    FROM 
        areas a
    LEFT JOIN 
        employers e ON a.id = e.area
    GROUP BY 
        a.name
),
region_vacancy_counts AS (
    SELECT 
        a.name AS region_name,
        COUNT(v.id) AS vacancy_count
    FROM 
        areas a
    LEFT JOIN 
        vacancies v ON a.id = v.area_id
    GROUP BY 
        a.name
)
SELECT 
    rec.region_name,
    rec.employer_count
FROM 
    region_employer_counts rec
JOIN 
    region_vacancy_counts rvc ON rec.region_name = rvc.region_name
WHERE 
    rvc.vacancy_count = 0
ORDER BY 
    rec.employer_count DESC
LIMIT 1;
'''

cur.execute(sql_query_2)
result_2 = cur.fetchone()

print("nРегион с наибольшим количеством работодателей без вакансий:")
if result_2:
    print(f"Регион: {result_2[0]}, Работодатели: {result_2[1]}")
else:
    print("Таких регионов нет.")

cur.close()
conn.close()

```

5. Напишите запрос, чтобы узнать название компании, находящейся на третьем месте в алфавитном списке (по названию) компаний, у которых указано четыре сферы деятельности. 

```python
# текст запроса
import psycopg2

conn = psycopg2.connect(
    dbname='project_sql',
    user='skillfactory',
    password='cCkxxLVrDE8EbvjueeMedPKt',
    host='84.201.134.129',
    port=5432
)

cur = conn.cursor()

# SQL-запрос для получения названия компании
sql_query = '''
WITH employer_industries_count AS (
    SELECT 
        e.id,
        e.name,
        COUNT(ei.industry_id) AS industry_count
    FROM 
        employers e
    LEFT JOIN 
        employers_industries ei ON e.id = ei.employer_id
    GROUP BY 
        e.id, e.name
)
SELECT 
    name
FROM 
    employer_industries_count
WHERE 
    industry_count = 4
ORDER BY 
    name
OFFSET 2
LIMIT 1;
'''

# Выполнение запроса
cur.execute(sql_query)
result = cur.fetchone()

# Вывод результата
print(f"Название компании: {result[0]}")

# Закрытие соединений
cur.close()
conn.close()
```

1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.

Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим.
